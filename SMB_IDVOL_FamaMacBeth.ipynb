```python
# Here I am able to import linearmodels
!pip install linearmodels
from linearmodels import FamaMacBeth
import pandas as pd
```

```python
df = pd.read_csv('/content/fm')
```

```python
pd.to_numeric(df['date'], errors = 'coerce')
pd.to_numeric(df['decile'], errors = 'coerce')
df = df.groupby(['decile','date']).mean()
df = df.dropna(0)
df
```

```python
import statsmodels.formula.api as smf
import numpy as np
```

```python
def ols_coef(x,formula):
    return smf.ols(formula,data=df).fit().params

gamma = (df.groupby('date')
  .apply(ols_coef,'decret ~ 1 + vwret + idvol'))

gamma.head()
```

```python
def fm_summary(p):
    s = p.describe().T
    s['std_error'] = s['std']/np.sqrt(s['count'])
    s['tstat'] = s['mean']/s['std_error']
    return s[['mean','std_error','tstat']]
```

```python
fm_summary(gamma)
```
